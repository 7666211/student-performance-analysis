```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set=pd.read_csv('Expanded_data_with_more_features.csv')
```

```python
print(data_set.head())
```

```python
data_set.describe()
```

```python
data_set.info()
```

```python
data_set.isnull().sum()
```

```python

#drope unnecessary column

data_set=data_set.drop("Unnamed: 0",axis=1)
```

```python
data_set.head()
```

# gender analysis

```python
plt.figure(figsize= (4,5))
graph = sns.countplot(data=data_set,x='Gender')
graph.bar_label(graph.containers[0])
plt.title('Gender distribution')
plt.show()
```

```python
#from the above analysis we learned that the number of females is more than that of male
```

# Parent education analysis

```python
gb=data_set.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
```

```python
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("relation between the parent's education and student's score")
plt.show()
```

```python
#from the above graph it seems that the parent education does matter on the student's score
```

# Parent marital status

```python
gb1=data_set.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot=True)
plt.title("relation between parent's marital status and student's score")
plt.show()
```

```python
#from the above graph it seems that the parent's marital status does not affect the student's score 
```

# distribution of Ethinic groups

```python
groupA=data_set.loc[(data_set['EthnicGroup']=="group A")].count()
groupB=data_set.loc[(data_set['EthnicGroup']=="group B")].count()
groupC=data_set.loc[(data_set['EthnicGroup']=="group C")].count()
groupD=data_set.loc[(data_set['EthnicGroup']=="group D")].count()
groupE=data_set.loc[(data_set['EthnicGroup']=="group E")].count()

label=['group A','group B','group C','group D','group E']
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist, labels=label,autopct='%1.2f%%')
plt.show()
```
